```python
#import sys
#!{sys.executable} -m pip install wordcloud
import os
import nltk
import numpy as np
import random
import string
import re
import glob
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt


#customstopwords = ["maybe", "right", "anything", "take", "time", "good", "okay", "know", "back", "need", "blood", "mean", "see", "say", "than", "oh", "think", "call", "people", "look", "come", "go", "us", "look", "victim", "going", "evidence", "know", "Hey", "uh", "guy", "something", "man", "house", "make", "Yeah", "find", "car", "want", "one", "tell", "Well", "Grissom", "thing"]

customstopwords = ["Grissom", "Amanda", "guy", "Hey", "going", "look", "Oh", "go", "know", "take", "Yeah", "Well", "say", "came", "Look", "Mr", "still", "said", "Nick", "Sara", "Catherine", "told", "guy", "even", "will", "looking", "two", "knew", "day", "put", "went", "will", "Warrick", "give", "first", "two", "three", "huh", "took", "lot", "let"]

#customstopwords = 

[STOPWORDS.add(n) for n in customstopwords]
stopwords = set(STOPWORDS) 

path = "/Users/dipes/Documents/Senior Year/Natural Language Processing/Final Project/data/character/good"

goodscript = ''
for filename in glob.glob(os.path.join(path, '*.txt')):
    input_file=open(filename, 'r')
    text=''
    for line in input_file:
        text += line
        goodscript += line
    
    characterfile=nltk.sent_tokenize(text)
    for i in range(len(characterfile)):
        characterfile[i]=characterfile[i].lower()
        characterfile[i]=re.sub(r'\W', ' ', characterfile[i])
        characterfile[i]=re.sub(r'\s+', ' ', characterfile[i])
    input_file.close()
    
path = "/Users/dipes/Documents/Senior Year/Natural Language Processing/Final Project/data/character/bad"

badscript = ''
for filename in glob.glob(os.path.join(path, '*.txt')):
    input_file=open(filename, 'r')
    text=''
    for line in input_file:
        text += line
        badscript += line
    
    characterfile=nltk.sent_tokenize(text)
    for i in range(len(characterfile)):
        characterfile[i]=characterfile[i].lower()
        characterfile[i]=re.sub(r'\W', ' ', characterfile[i])
        characterfile[i]=re.sub(r'\s+', ' ', characterfile[i])
    input_file.close()


def show_wordcloud(script, title = None):
    wordcloud = WordCloud(background_color='white', stopwords=stopwords, max_words=50,max_font_size=40, scale=3, random_state=1).generate(str(script))
    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()

show_wordcloud(goodscript)
show_wordcloud(badscript)
```
